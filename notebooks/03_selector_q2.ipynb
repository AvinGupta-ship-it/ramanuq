## Gate V4 — exact recovery on rigged cases (run before any real number)

```python
import numpy as np
import pandas as pd

from ramanuq.selectors import (
    audit,
    coverage_under_misspecification,
    rigged_cases,
)

_EPS = 1e-12
_correlated, _anti = rigged_cases()

_t6_corr = audit(_correlated)
_t6_anti = audit(_anti)

assert np.allclose(_t6_corr["rho_median"], 1.0, rtol=0.0, atol=_EPS), "V4 rho(+1)"
assert np.allclose(_t6_corr["regret_median"], 0.0, rtol=0.0, atol=_EPS), "V4 regret 0"
assert np.allclose(_t6_anti["rho_median"], -1.0, rtol=0.0, atol=_EPS), "V4 rho(-1)"
print("Gate V4 PASSED: rho=+1/regret=0 on correlated, rho=-1 on anti-correlated.")
```

## F5 (deferred to Day 9 — built with viz.py)

F5 is a Day-9 deliverable built with `viz.py`; it is NOT rendered here. The cell
below only assembles the arrays F5 will plot (rho point estimate and CI band per
selector/stratum). No figure is produced and `viz` is not imported.

```python
# Arrays F5 will consume on Day 9 (stored, not plotted).
f5_arrays = {}
for (stratum, selector), cell in t6.groupby(["stratum", "selector"]):
    cell = cell.sort_values("snr_label")
    f5_arrays[(stratum, selector)] = {
        "snr_label": cell["snr_label"].to_numpy(),
        "rho_median": cell["rho_median"].to_numpy(),
        "rho_lo": cell["rho_lo"].to_numpy(),
        "rho_hi": cell["rho_hi"].to_numpy(),
        "hit_rate": cell["hit_rate"].to_numpy(),
    }
list(f5_arrays.keys())
```
